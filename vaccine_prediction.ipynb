```python
X = df.iloc[:,:-1]
y = df.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
sl = StandardScaler()
df1_sl = sl.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(df1_sl,y, test_size=0.3, random_state=3)
```

```python
from sklearn import svm, tree, linear_model, neighbors, naive_bayes, ensemble, discriminant_analysis, gaussian_process
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix,plot_confusion_matrix,plot_roc_curve

MLA = [                                                #Creating a list of All Classification Algorithms
       linear_model.LogisticRegression(),
       ensemble.AdaBoostClassifier(),
       ensemble.BaggingClassifier(),
       ensemble.GradientBoostingClassifier(),
       ensemble.RandomForestClassifier(), 
       linear_model.SGDClassifier(),
       XGBClassifier(),
       naive_bayes.GaussianNB(),
       neighbors.KNeighborsClassifier(),
       tree.DecisionTreeClassifier(),
       tree.ExtraTreeClassifier(),
       svm.SVC() ]
    
    
```

```python
name = []
testscore = []
trainscore = []
for alg in MLA:
    name.append(alg.__class__.__name__)
    alg.fit(X_train, y_train)
    test_pred = alg.predict(X_test)
    testscore.append(accuracy_score(y_test,test_pred))
    train_pred = alg.predict(X_train)
    trainscore.append(accuracy_score(y_train, train_pred))
    print(alg.__class__.__name__)
    print('__________________________________')
    print(classification_report(y_test,test_pred))   
    print(confusion_matrix(y_test,test_pred))
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms_df = mms.fit_transform(X)
mms_data = pd.DataFrame(mms_df, columns = X.columns)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95,random_state=45)
pca_data = pca.fit_transform(mms_data)
```

```python
pd.Series(zip(list(pca.explained_variance_ratio_), list(np.cumsum(pca.explained_variance_ratio_))))
```

```python
final = pd.DataFrame(pca_data, columns = ['PCA' + str(i) for i in range(1,27)])
final.head()
```

```python
X = final
y = y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.4, random_state = 10)
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(X_train,y_train)
y_pred=model1.predict(X_test)
y_tr_pr = model1.predict(X_train)
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
